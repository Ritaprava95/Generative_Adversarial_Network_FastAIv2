## From these images we need to make pytorch dataloaders

```python
from fastai.vision.all import *
from fastai.basics import *
from fastai.callback.all import *
from fastai.vision import gan
from fastai.vision.gan import *
import torch
import fastai
```

```python
print(torch.__version__)
print(fastai)
```

```python
training_img_path = r"E:\Work\Data_Science\Projects\Custom_GAN\data\train\house"
```

```python
dblock = DataBlock(blocks = (TransformBlock, ImageBlock),
                   get_x = generate_noise,
                   get_items = get_image_files,
                   splitter = IndexSplitter([]),
                   item_tfms=Resize(size=64, method=ResizeMethod.Crop), 
#                    batch_tfms = Normalize.from_stats(torch.tensor([0.5,0.5,0.5]), torch.tensor([0.5,0.5,0.5]))
                  )

dls = dblock.dataloaders(training_img_path, bs=8)
```

```python
dls.show_batch(max_n=4)
```

### Save the dls
